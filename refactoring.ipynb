## 🎯 Our task: calculate the Manhattan distance

<img alt="Manhattan vs Euclidean distance" src="https://drive.google.com/thumbnail?id=1yFNtDikccITMCMXUL0kxIgcCSVwbEOlu&sz=w1000">

* **Öklid mesafesi**: Uzayda iki nokta arasındaki düz çizgi mesafesi.
  Formül:

$$
  \sqrt{(x_2 - x_1)^2 + (y_2 - y_1)^2}
  $$

* **Manhattan mesafesi**: Koordinatların mutlak farklarının toplamı (bir ızgara üzerinde gezinmek gibi).
  Formül:

$$
|x_2 - x_1| + |y_2 - y_1|
$$

First, we 📝 draft some code in a notebook cell:

````python
a = (1, 1)
b = (4, 5)

d_x = b[0] - a[0]
d_y = b[1] - a[1]

distance = d_x + d_y
distance
````

(Evet, bu kodda bir sorun var, ama görmezden gelelim.)

Güzel görünüyor! Ama bu kodu yeniden kullanamayız.

Bunu, iki nokta, yani iki tuple'ı argüman olarak alacak olan `manhattan(a, b)` işlevine yeniden düzenleyelim:

````python
def manhattan(a, b):
    d_x = b[0] - a[0]
    d_y = b[1] - a[1]
    return abs(d_x) + abs(d_y)
````

Fonksiyonunuzu deneyin:

````python
a = (1, 1)
b = (4, 5)
manhattan(a, b)
````

## Refactoring

İki fonksiyonumuz birbirine çok benziyor.

Ortak kısımları yeniden kullanmak için bunları [**yeniden düzenleyebiliriz**](https://en.wikipedia.org/wiki/Code_refactoring).

Manhattan (`p = 1`) ve Öklid (`p = 2`) mesafeleri, [**Minkowski**](https://en.wikipedia.org/wiki/Minkowski_distance) mesafesinin özel durumlarıdır:

$$AB = \sqrt[\raisebox{16mu}{\hspace{-2mu}$\scriptstyle\frac1p$}]{(\color {red}{b_x - a_x} \color {black})^p + (\color {teal}{b_y - a_y}) \color {black}^p}$$

Minkowski'yi kodlayalım ve bu fonksiyonu Manhattan ve Öklid hesaplamaları için yeniden kullanalım!

İşte ilk (yanlış) denememiz:

```python
def manhattan(a, b):
    return minkowski(a, b, 1)


def euclidean(a, b):
    return minkowski(a, b, 2)


def minkowski(a, b, p):
    d_x = b[0] - a[0]
    d_y = b[1] - a[1]

    distance = (d_x**p + d_y**p)**(1/p)
    return distance
```

Testleri tekrar çalıştırın, kodu düzeltin:

```python
    distance = (abs(d_x)**p + abs(d_y)**p)**(1/p)
```
